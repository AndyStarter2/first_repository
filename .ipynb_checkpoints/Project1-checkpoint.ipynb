```python
import pandas as pd

# Создаём датафрейм
data = {
  'Модель': ['Dodge RAM',
             'Chevrolet Tahoe',
             'Cadillac Escalade',
             'Ford Mustang',
             'Chevrolet Comaro',
             'Dodge Challenger'],
  'Тип кузова': ['SUV', 'SUV', 'SUV', 'sedan', 'sedan', 'sedan'],
  'Масса (тонны)': [2.5, 2.9, None, 1.7, None, 2]
  }
df = pd.DataFrame(data)

# Вычисляем среднюю массу автомобилей с кузовом SUV
mean_suv = df[df['Тип кузова'] == 'SUV']['Масса (тонны)'].mean()
# Вычисляем среднюю массу автомобилей с кузовом sedan
mean_sedan = df[df['Тип кузова'] == 'sedan']['Масса (тонны)'].mean()

# Заполняем пропуски
def weight_by_type(x):
    if pd.isna(x['Масса (тонны)']):
        if x['Тип кузова'] == 'SUV':
            return mean_suv
        else:
            return mean_sedan
    else:
        return x['Масса (тонны)']

# Заполните пропуски, используя функцию weight_by_type()
df['Масса (тонны)'] = df.apply(weight_by_type, axis=1)

# Выводим результат
print(df)
```

```python
# Импортируем библиотеку pandas
import pandas as pd
# Выгружаем данные в датафрейм
df = pd.read_csv('https://code.s3.yandex.net/datasets/hotel_dataset.csv')
# Сохраняем общее количество строк датафрейма
df_lines = df.shape[0]

# Напишите ваш код здесь
df = df.dropna(subset=['hotel Name', 'hotel Address', 'price for the night'])

# Сохраняем количество удалённых строк
deleted_lines = df_lines - df.shape[0]
# Сохраняем процент удалённых строк
deleted_percent = round(deleted_lines / df_lines * 100, 2)
# Выводим результат
print(f'Удалено всего {deleted_percent}% или {deleted_lines} строк.')
```
